```python
# Show a plot of the response rate by the sales channel.
# I created a function to calc the response rate (number of Yes / number of total for each sales channel) and stored them in a dict
# After I transform the dictonary into an dataframe and plot the result with sns.barplot
# I also found the function groupby

def calcResponseRate(dataframe, index, searching_row):
    count = 0
    result_dict = {}
    for row in index:
        row_value = row
        yes_value = sum((dataframe[searching_row] == row) & (df['Response'] == 'Yes'))
        no_value = sum((dataframe[searching_row] == row) & (df['Response'] == 'No'))
        l = (yes_value / (yes_value + no_value)) * 100
        result_dict[count] = [row_value, round(l, 2)]
        count += 1
    df_v = pd.DataFrame.from_dict(result_dict).T
    df_v.columns = [searching_row, 'Response']
    df_v
    return df_v


    
    
sales_channels = df['Sales Channel'].unique()

z = calcResponseRate(df, sales_channels, 'Sales Channel')

    
print(z)



```

```python

sns.barplot(x="Sales Channel", y='Response', data = z)
plt.title('Response rates for each Sales Channel')
plt.ylabel('Response rate in %')
plt.show()
```

```python
# https://www.youtube.com/watch?v=ipoSjrN0oh0 

# want to split the different total claim amount in different groups to show different response rates for each group
bins = pd.IntervalIndex.from_tuples([(0,499), (500,999), (1000,1499), (1500,1999), (2000,2499), (2500, 2999)])

df['total claim bins'] = pd.cut(df['Total Claim Amount'], bins)

display(df.groupby(['total claim bins'])['Total Claim Amount'].agg(['min','max']))


```

```python
# tried to use hue for splitting the plot and also using countplot

display(df)

sns.countplot(x='total claim bins', hue = 'Response', data=df)
plt.title('Response Rate nach Total Claim Amount')
plt.xlabel('Total Claim Amount')
plt.ylabel('Response Rate')
plt.show()
```

```python
yes_df = df[(df['Response'] == 'Yes')]
claim_response_rate = round((yes_df.groupby(['total claim bins'])['Response'].count() ) / (df.groupby(['total claim bins'])['Response'].count()) * 100, 2)
temp = pd.DataFrame(claim_response_rate)
temp = temp.reset_index()
temp.columns = ['Range of total claim Amount', 'Response Rate']
display(temp)


sns.barplot(x='Range of total claim Amount', y='Response Rate', data=temp)
plt.title('Response rates for groups of total claim amount')
plt.ylabel('Response rate in %')
plt.xlabel('Groups of total claim amount (range = 500)')
plt.show
```
